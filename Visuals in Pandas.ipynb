# Pandas Visualizations 

```python
import pandas as pd 
import numpy as mp 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv(r'/Users/samwekshakya/Downloads/Ice Cream Ratings.csv')
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('seaborn-v0_8-pastel')
```

# Line PLot

```python
df.plot(kind = 'line', subplots = True)
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel='Daily Ratings', ylabel='Scores')
```

# Bar Plot

```python
df.plot(kind='bar')
```

```python
df.plot(kind='bar', stacked= True)
```

```python
df['Flavor Rating'].plot(kind='bar', stacked= True)
```

```python
df.plot.barh(stacked= True)
```

# Scatter Plot

```python
df.plot.scatter(x='Texture Rating', y= 'Overall Rating', s=500, c='Cyan')
```

# Histogram

```python
df.plot.hist(bins=20)
```

# Box Plot 

```python
df.boxplot()
```

# Area Plot

```python
df.plot.area(figsize=(10,5))
```

# Pie Chart

```python
df.plot.pie(y='Flavor Rating',figsize=(10,5))
```
